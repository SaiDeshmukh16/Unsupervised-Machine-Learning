```python
import pandas as pd
import seaborn as sns
```

```python
url = "https://drive.google.com/file/d/1oYQSNxfvw6kFr6-N9rKLRAnLXlp0osEt/view?usp=sharing"
path = "https://drive.google.com/uc?export=download&id="+url.split("/")[-2]
df_10 = pd.read_csv(path) 
df_10
```

```python
df_10_copy = df_10.copy()
df_10_copy
```

```python
df_10_copy.drop(df_10_copy.columns[[4, 6, 13, 14, 15, 16, 17]], axis=1, inplace=True)
```

```python
df_10_copy.rename(columns={df_10_copy.columns[0]: 'name'},inplace=True)
df_10_copy.rename(columns={df_10_copy.columns[1]: 'artist'},inplace=True)
df_10_copy.rename(columns={df_10_copy.columns[2]: 'danceability'},inplace=True)
df_10_copy.rename(columns={df_10_copy.columns[3]: 'energy'},inplace=True)
df_10_copy.rename(columns={df_10_copy.columns[4]: 'loudness'},inplace=True)
df_10_copy.rename(columns={df_10_copy.columns[5]: 'speechiness'},inplace=True)
df_10_copy.rename(columns={df_10_copy.columns[6]: 'acousticness'},inplace=True)
df_10_copy.rename(columns={df_10_copy.columns[7]: 'instrumentalness'},inplace=True)
df_10_copy.rename(columns={df_10_copy.columns[8]: 'liveness'},inplace=True)
df_10_copy.rename(columns={df_10_copy.columns[9]: 'valence'},inplace=True)
df_10_copy.rename(columns={df_10_copy.columns[10]: 'tempo'},inplace=True)
print(df_10_copy.columns)
# df_10_copy
```

```python
df_10_copy = df_10_copy.set_index(["name","artist"])
```

```python
# 1. import the transformer
from sklearn.preprocessing import StandardScaler

# 2. initialize the transformer 
my_std = StandardScaler()

#3. Fit the transformer to the data
my_std.fit(df_10_copy)

#4. use the transfomer to transform the data
scaled_features = my_std.transform(df_10_copy)
#5. reconvert the transformed dta back to a Data Frame
scaled_features_df_10 = pd.DataFrame(scaled_features, 
                                  index=df_10_copy.index, 
                                  columns=df_10_copy.columns)
```

```python
# 1. import the model
from sklearn.cluster import KMeans
 
# 2. initialize the model
my_kmeans = KMeans(n_clusters= 21, random_state = 123) # you always choose the number of k here
 
# 3. fit the model to the data
my_kmeans.fit(scaled_features_df_10) #pass your scaled data here

#4.1 obtain the cluster output (no need to use this one)
#clusters = my_kmeans.predict(scaled_features_df_10) # pass your scaled data here

# 4.2 obtain the cluster output
clusters = my_kmeans.labels_
 
 # 5. attach the cluster output to our original DataFrame
scaled_features_df_10_with_clusters = scaled_features_df_10.copy()
scaled_features_df_10_with_clusters["cluster"] = clusters

scaled_features_df_10_with_clusters
# scaled_features_df_10
# clusters
```

```python
my_kmeans.inertia_
```

```python
my_kmeans.inertia_
max_k = 20

inertia_list = []

for i in range(1, max_k):
    myKMeans = KMeans(n_clusters=i, random_state = 123)
    myKMeans.fit(scaled_features_df_10)
    inertia_list.append(round(myKMeans.inertia_))
```

```python
sns.set_theme(style='darkgrid')

(
sns.relplot(x=range(1, max_k),
           y=inertia_list, 
           kind = "line",
           marker='o',
           height=8,
           aspect=2)
.set(title=f"Inertia score from a to {max_k} clusters")
.set_axis_labels("Number of clusters", "Inertia score")
);
```

```python
from sklearn.metrics import silhouette_score

silhouette_score(scaled_features_df_10, my_kmeans.labels_)
```

```python
max_k = 30
sil_score = []

for k in range(2, max_k):
    #prepare a kmeans with i-clusters
    myKMeans = KMeans(n_clusters=k, random_state = 123)
    #apply Kmeans on our scaled features
    myKMeans.fit(scaled_features_df_10)
    sil_score.append(silhouette_score(scaled_features_df_10, myKMeans.labels_))
```

```python
(
sns.relplot(x=range(2, max_k),
           y=sil_score, 
           kind = "line",
           marker='o',
           height=8,
           aspect=2)
.set(title=f"Silhouette score from 2 to {max_k} clusters")
.set_axis_labels("Number of clusters", "Silhouette score")
);
```
